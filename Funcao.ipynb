```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plotar_funcao(funcao, intervalo, titulo, cor="r", estilo="-"):
  # Gerando pontos no intervalo
  x = np.linspace(intervalo[0], intervalo[1], 100)
  y = [funcao(valor) for valor in x]

  # Plotando a função
  plt.plot(x, y, color=cor, linestyle=estilo)
  plt.title(titulo)
  plt.xlabel("x")
  plt.ylabel("f(x)")
  plt.grid(True)
```

```python
# Exemplo de uso
def f(x):
  return x**2
```

```python
plotar_funcao(f, (-5, 5), "Gráfico da função f(x) = x^2")

# Mostrando o gráfico
plt.show()
```
