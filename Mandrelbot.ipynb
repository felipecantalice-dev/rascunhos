```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def mandelbrot(c, max_iter=80):
    z = 0
    n = 0
    while abs(z) <= 2 and n < max_iter:
        z = z*z + c
        n += 1
    return n

# Create a grid of complex numbers
xmin, xmax = -2.25, 0.75
ymin, ymax = -1.5, 1.5
width, height = 1000, 1000
real_vals = np.linspace(xmin, xmax, width)
imag_vals = np.linspace(ymin, ymax, height)

# Calculate Mandelbrot membership for each point
data = np.empty((width, height))
for i, x in enumerate(real_vals):
    for j, y in enumerate(imag_vals):
        c = complex(x, y)
        data[i, j] = mandelbrot(c)

plt.imshow(data, cmap='hot', interpolation='nearest')
plt.axis('off')
plt.show()
```
