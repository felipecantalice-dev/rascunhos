```python
import random
import numpy as np

recompensa_cooperar = 3
recompensa_desertar = 5
punição_desertar = 1
tentação_desertar = 2


def jogar_rodada(jogador1_escolha, jogador2_escolha):
  if jogador1_escolha and jogador2_escolha:
    return recompensa_cooperar, recompensa_cooperar
  elif jogador1_escolha and not jogador2_escolha:
    return recompensa_desertar, punição_desertar
  elif not jogador1_escolha and jogador2_escolha:
    return punição_desertar, recompensa_desertar
  else:
    return tentação_desertar, tentação_desertar


numero_rodadas = 1000

# 0 para sempre cooperar
# 1 para sempre desertar
# 2 para aleatório

jogador1_estrategia = 0
jogador2_estrategia = 1

jogador1_pontos = 0
jogador2_pontos = 0

for _ in range(numero_rodadas):
  if jogador1_estrategia == 0:
    jogador1_escolha = True
  elif jogador1_estrategia == 1:
    jogador1_escolha = False
  else:
    jogador1_escolha = random.choice([True, False])

  if jogador2_estrategia == 0:
    jogador2_escolha = True
  elif jogador2_estrategia == 1:
    jogador2_escolha = False
  else:
    jogador2_escolha = random.choice([True, False])

  pontos_jogador1, pontos_jogador2 = jogar_rodada(jogador1_escolha, jogador2_escolha)
  jogador1_pontos += pontos_jogador1
  jogador2_pontos += pontos_jogador2

print(f"Pontos Jogador 1: {jogador1_pontos}")
print(f"Pontos Jogador 2: {jogador2_pontos}")
```
