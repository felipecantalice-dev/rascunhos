```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lista_numeros_complexos = [
    complex(-0.8, 0.156),
    complex(-0.4, -0.1428j),
    complex(0.390544, 0.587496j),
    complex(-0.835, -0.2321j),
    complex(-0.70176, -0.3842j),
]
```

```python
c = lista_numeros_complexos[0]

largura, altura = 500, 500

# Define os limites da coordenada x.
xmin = -2
xmax = 1

# Define os limites da coordenada y.
ymin = -1.5
ymax = 1.5
```

```python
julia_set = np.zeros((largura, altura), dtype=np.int32)

plt.imshow(julia_set, cmap="hot")
plt.show()
```

```python
# f(z)=z^2+c
def f(z,c):
  return z**2+c

# Percorre cada pixel
for i in range(largura):
  for j in range(altura):

    # Ajusta os pixels
    x = xmin + (xmax - xmin) * i / largura
    y = ymin + (ymax - ymin) * j / altura
    # Cria um valor complexo
    z = complex(x,y)
    # Quantidade de interacoes
    for n in range(256):
      # Se nao estiver dentro do conjunto de Julia
      if abs(z) > 3:
        break

      z = f(z,c)

    julia_set[i,j] = n

# Plota o conjunto de Julia.
plt.imshow(julia_set, cmap="hot")
plt.show()
```
