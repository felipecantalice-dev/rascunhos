# Descida do gradiente

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np


def f(x, y):
    """
    Function to be plotted and evaluated.

    Args:
        x: x-coordinate value.
        y: y-coordinate value.

    Returns:
        The value of f(x, y).
    """
    return x**2 * np.sin(y)


# Define the range for the plot
x_min, x_max = -3, 3
y_min, y_max = -3, 3

# Create a 3D grid of points
x, y = np.meshgrid(np.linspace(x_min, x_max, 200), np.linspace(y_min, y_max, 200))

# Calculate the values of f(x, y) at each point
z = f(x, y)

# Create a 3D plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(x, y, z, cmap='viridis')

# Set axis labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(x, y)')
ax.set_title('Plot of f(x, y) = x^2 * sin(y)')

# Set colorbar
fig.colorbar(surf, label='f(x, y)')

# Set viewing angles
ax.view_init(elev=15, azim=-60)

# Show the plot
plt.show()

# Calculate and print f(2, -2)
result = f(2, -2)
print("f(2, -2) =", result)
```
