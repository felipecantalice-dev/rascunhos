```python
def armstrong(number: int) -> bool:
  temp = number
  soma = 0

  while temp != 0:
    resto = temp%10
    soma += resto**3
    temp= temp//10

  if soma == number:
    return True
  return False
```

```python
lista = []

for i in range(1700):
  isArmstrong = armstrong(i)
  if isArmstrong:
    lista.append(i)

lista
```
