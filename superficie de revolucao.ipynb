```python
# Importar as bibliotecas
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm

# Criar uma figura e um eixo 3D
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# Criar uma malha de pontos (X, Y, Z)
theta = np.linspace(0, 2 * np.pi, 50) # ngulo de rotao
r = np.linspace(-5, 5, 50) # raio da curva
R, Theta = np.meshgrid(r, theta) # malha polar
X = R * np.cos(Theta) # coordenada x
Y = R * np.sin(Theta) # coordenada y
Z = -X**2 # coordenada z

# Plotar a superfcie de revoluo
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm)

# Ajustar os limites e os rtulos dos eixos
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_zlim(-25, 0)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

# Mostrar o grfico
plt.show()
```

```python
# Importar as bibliotecas necessrias
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Criar a figura e o eixo 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Definir os limites dos parmetros
u_min = 0
u_max = 2 * np.pi
v_min = 0
v_max = 2 * np.pi

# Criar os vetores u e v
u = np.linspace(u_min, u_max, 50)
v = np.linspace(v_min, v_max, 50)

# Criar as matrizes U, V e as funes X, Y e Z
U, V = np.meshgrid(u, v)
X = (2 + np.cos(V)) * np.cos(U)
Y = (2 + np.cos(V)) * np.sin(U)
Z = np.sin(V)

# Plotar a superfcie
ax.plot_surface(X, Y, Z, cmap='viridis')

# Definir os rtulos dos eixos
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

# Mostrar o grfico
plt.show()
```
