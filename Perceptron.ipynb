```python
import numpy as np

class Perceptron:

    def __init__(self, taxa_aprendizado=0.1, epocas=100):
        self.taxa_aprendizado = taxa_aprendizado
        self.epocas = epocas
        self.pesos = None
        self.bias = None

    def treinar(self, entradas, saidas):
        # Inicializar pesos e bias
        self.pesos = np.random.rand(entradas.shape[1])
        self.bias = np.random.rand()

        for _ in range(self.epocas):
            for i in range(len(entradas)):
                # Calcular a saída da rede
                u = np.dot(entradas[i], self.pesos) + self.bias
                y = self.sinal(u)

                # Atualizar pesos e bias
                erro = saidas[i] - y
                self.pesos += self.taxa_aprendizado * erro * entradas[i]
                self.bias += self.taxa_aprendizado * erro
        print(f'{self.pesos}, {self.bias}')

    def testar(self, entrada):
        # Calcular a saída da rede
        u = np.dot(entrada, self.pesos) + self.bias
        y = self.sinal(u)

        return y

    def sinal(self, u):
        return 1 if u >= 0 else 0
```

```python
# Conjunto de dados
entradas = np.array([
    [5, 3],
    [4, 2],
    [7, 4],
    [6, 5],
])

saidas = np.array([0, 0, 1, 1])

# Treinar o Perceptron
perceptron = Perceptron()
perceptron.treinar(entradas, saidas)

# Testar o Perceptron
nova_flor = np.array([5, 3.5])
resultado = perceptron.testar(nova_flor)

# Imprimir o resultado
if resultado == 0:
    print("A flor é provavelmente uma rosa.")
else:
    print("A flor é provavelmente um lírio.")
```

```python
import matplotlib.pyplot as plt

# Definir cores para as classes
cores = {
    0: "r", # Rosa
    1: "b", # Lírio
}

# Plotar os pontos
for i in range(len(entradas)):
    plt.scatter(entradas[i][0], entradas[i][1], color=cores[saidas[i]])

plt.scatter(nova_flor[0], nova_flor[1], color='g')

# Adicionar rótulos aos eixos
plt.xlabel("Comprimento da pétala")
plt.ylabel("Largura da pétala")

# Mostrar o gráfico
plt.show()
```
