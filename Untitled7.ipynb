```python
molecula = "SoC.CaO"

# Divide a string em uma lista de substrings usando o ponto como separador
molecula_dividida = molecula.split(".")
molecula_dividida
```

```python
hash = {}

for molecula in molecula_dividida:

  letter = ""

  for i in range(0, len(molecula) -1):

    if molecula[i].isupper() and molecula[i+1].islower():
      letter = molecula[i] + molecula[i+1]

      hash[letter] = hash[letter] + 1 if hash[letter].get() else 0
      i = i + 2
      letter = ""

hash
```
