# Sequencia de Lucas
sequência de números inteiros, começando com 2 e 1, em que cada número subsequente é a soma dos dois números anteriores.

## Fórmula
$L(n) = L(n-1) + L(n-2)$

* Os primeiros termos da sequência de Lucas são: 2, 1, 3, 4, 7, 11, 18, 29, 47, 76, ...
* Cada termo da sequência é a soma dos dois termos anteriores.
* A sequência de Lucas é semelhante à sequência de Fibonacci, mas começa com 2 e 1 em vez de 0 e 1.


```python
def lucas(n):
  """
  Função para calcular a sequência de Lucas.

  Argumentos:
    n: Número de termos da sequência a serem gerados.

  Retorno:
    Lista com os n primeiros termos da sequência de Lucas.
  """
  if n == 0:
    return []
  elif n == 1:
    return [2]
  elif n == 2:
    return [2, 1]
  else:
    luc = [2, 1]
    for i in range(2, n):
      luc.append(luc[i-1] + luc[i-2])
    return luc

def plot_lucas(luc):
  """
  Função para plotar a sequência de Lucas.

  Argumentos:
    luc: Lista com os termos da sequência de Lucas.

  Retorno:
    None. Plota a sequência de Lucas em um gráfico de linhas.
  """
  import matplotlib.pyplot as plt

  plt.plot(range(len(luc)), luc)
  plt.xlabel("Termo")
  plt.ylabel("Valor")
  plt.title("Sequência de Lucas")
  plt.show()

# Gerando os 20 primeiros termos da sequência
n = 20
luc = lucas(n)

# Exibindo os termos da sequência
print("Sequência de Lucas:")
print(luc)

# Plotando a sequência
plot_lucas(luc)
```
