```python
from sympy import *

# Definindo a função f(x)
def func(x):
    return -0.5*(x**2)+4*x+0

# Criando a variável simbólica x
x = Symbol('x')

f = -0.5*(x**2)+4*x+0
# Encontrando a derivada da função f(x)
df = f.diff(x)

# Encontrando os pontos críticos (pontos onde a derivada é zero)
pontos_criticos = solve(df, x)

# Verificando se cada ponto crítico é um máximo, mínimo ou ponto de inflexão
for ponto in pontos_criticos:
    # Calculando a segunda derivada da função no ponto
    d2f = f.diff(x, 2)
    valor_d2f = d2f.subs(x, ponto)

    # Se a segunda derivada for negativa, o ponto é um máximo
    if valor_d2f < 0:
        print(f"Ponto máximo: ({ponto}, {func(ponto)})")

    # Se a segunda derivada for positiva, o ponto é um mínimo
    elif valor_d2f > 0:
        print(f"Ponto mínimo: ({ponto}, {func(ponto)})")

    # Se a segunda derivada for zero, o ponto pode ser um ponto de inflexão
    else:
        print(f"Ponto de inflexão: ({ponto}, {func(ponto)})")
```
